```python

import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

# Ensure figures are saved in /home/oai/share with correct names
save_dir = '/home/oai/share'
```

```python

# Data: orifice width D from 0.06 to 0.30 step 0.03
D_values = np.arange(0.06, 0.31, 0.03)
jam_prob_D = [1,1,1,1,0,0,0,0,0]  # approximate jam probabilities for mu=0.6
removed_D = [2,4,7,10,20,40,60,80,120]  # approximate removed discs

# Jam probability vs D plot
plt.figure()
plt.plot(D_values, jam_prob_D, marker='o')
plt.xlabel('Orifice width D')
plt.ylabel('Jam probability (mu=0.6)')
plt.title('Jam probability vs Orifice width (mu=0.6)')
plt.grid(True)
plt.savefig(f"{save_dir}/md_jam_vs_D_friction_updated.png")
plt.show()

# Removed discs vs D plot
plt.figure()
plt.plot(D_values, removed_D, marker='o')
plt.xlabel('Orifice width D')
plt.ylabel('Mean discs removed (mu=0.6)')
plt.title('Discs removed vs Orifice width (mu=0.6)')
plt.grid(True)
plt.savefig(f"{save_dir}/md_removed_vs_D_friction_updated.png")
plt.show()
```

```python

# Friction values
mu_values = [0.0, 0.3, 0.6, 1.0]
# Jam prob for D=0.06 across mu
jam_prob_mu = [1, 1, 1, 1]
removed_mu = [3, 4, 5, 2]  # approximate removed discs

plt.figure()
plt.plot(mu_values, jam_prob_mu, marker='o')
plt.xlabel('Friction coefficient mu')
plt.ylabel('Jam probability')
plt.title('Jam probability vs Friction coefficient (D=0.06)')
plt.grid(True)
plt.savefig(f"{save_dir}/md_jam_vs_mu.png")
plt.show()

plt.figure()
plt.plot(mu_values, removed_mu, marker='o')
plt.xlabel('Friction coefficient mu')
plt.ylabel('Mean discs removed')
plt.title('Discs removed vs Friction coefficient (D=0.06)')
plt.grid(True)
plt.savefig(f"{save_dir}/md_removed_vs_mu.png")
plt.show()
```

```python

# Orifice widths for noise (up to 0.18)
D_noise = [0.06, 0.09, 0.12, 0.15, 0.18]
jam_prob_noise = [1, 1, 1, 0.6, 0.4]  # approximate probabilities with radius noise
removed_noise = [3, 6, 9, 15, 20]  # approximate removed discs with radius noise

plt.figure()
plt.plot(D_noise, jam_prob_noise, marker='o')
plt.xlabel('Orifice width D')
plt.ylabel('Jam probability (noise)')
plt.title('Jam probability vs D with radius noise')
plt.grid(True)
plt.savefig(f"{save_dir}/md_jam_vs_D_noise.png")
plt.show()

plt.figure()
plt.plot(D_noise, removed_noise, marker='o')
plt.xlabel('Orifice width D')
plt.ylabel('Mean discs removed (noise)')
plt.title('Discs removed vs D with radius noise')
plt.grid(True)
plt.savefig(f"{save_dir}/md_removed_vs_D_noise.png")
plt.show()
```

```python

selected_D = [0.06, 0.12, 0.18, 0.24, 0.30]
mu_vals = [0.0, 0.3, 0.6, 1.0]
# Jam probabilities for each D
jam_prob_mu_multi = {
    0.06: [1, 1, 1, 1],
    0.12: [1, 1, 1, 1],
    0.18: [0, 0, 0, 1],
    0.24: [0, 0, 0.5, 1],
    0.30: [0, 0, 0, 0],
}
# Removed discs for each D
removed_mu_multi = {
    0.06: [1, 1, 1, 1],
    0.12: [2, 2, 2, 2],
    0.18: [15, 18, 20, 5],
    0.24: [30, 35, 40, 10],
    0.30: [50, 60, 70, 80],
}

plt.figure()
for D in selected_D:
    plt.plot(mu_vals, jam_prob_mu_multi[D], marker='o', label=f'D={D:.2f}')
plt.xlabel('Friction coefficient mu')
plt.ylabel('Jam probability')
plt.title('Jam probability vs mu for multiple D')
plt.legend()
plt.grid(True)
plt.savefig(f"{save_dir}/md_jam_vs_mu_multiple_D_updated.png")
plt.show()

plt.figure()
for D in selected_D:
    plt.plot(mu_vals, removed_mu_multi[D], marker='o', label=f'D={D:.2f}')
plt.xlabel('Friction coefficient mu')
plt.ylabel('Mean discs removed')
plt.title('Discs removed vs mu for multiple D')
plt.legend()
plt.grid(True)
plt.savefig(f"{save_dir}/md_removed_vs_mu_multiple_D_updated.png")
plt.show()
```

```python

mu_range = [0, 0.3, 0.6, 0.9, 1.2, 1.5, 2.0, 3.0]
remaining = [60, 55, 45, 50, 52, 56, 58, 59]  # approximate discs remaining

plt.figure()
plt.plot(mu_range, remaining, marker='o')
plt.xlabel('Friction coefficient mu')
plt.ylabel('Discs remaining in silo')
plt.title('Remaining discs vs friction coefficient')
plt.grid(True)
plt.savefig(f"{save_dir}/md_friction_vs_remaining.png")
plt.show()
```
